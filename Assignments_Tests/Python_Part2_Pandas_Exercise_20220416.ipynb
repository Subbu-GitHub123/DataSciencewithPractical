### 1. Find the most expensive car company name

```python
df[:10]
```

```python
highest_price = max(df['price'])
make_of_most_expensive_car = df[df['price'] == highest_price]['make'].to_string(index=False).strip()
print("Most expensive car is: ", make_of_most_expensive_car)
```

### 3. Find each company’s highest price car

```python
car_manufacturers = df.groupby('make')
priceDf = car_manufacturers['make','price'].max()
priceDf
```

### 4. Sort all cars by Price column

```python
df_sorted = df.sort_values(by=['price'], ascending=True)
df_sorted.head(5)
```

### 5. Group by make, fuel-type and body-style to calculate the average of city-mpg and highway-mpg (miles per gallon)

```python
df_groupby = df.groupby(['make', 'fuel_type', 'body_style'])
df_avg = df_groupby['city_mpg','highway_mpg'].mean()
df_avg
```
